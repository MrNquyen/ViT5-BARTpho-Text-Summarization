```python
import os

os.chdir(r"/datastore/npl/ViInfographicCaps/workspace/baseline_summarization/vit5_summarize/Seq2SeqTransformer-Text-Summarization")

from utils.utils import load_json, save_json
```

```python
split_path_template = "/datastore/npl/ViInfographicCaps/data/data/vitern/train_dev_test/{split}.json"
all_splits = ["train", "val", "test"]
```

# **PROCESS**

```python
save_path_template = "/datastore/npl/ViInfographicCaps/data/data/vitern/train_dev_test/{split}_no_segmentation.json"
for split in all_splits:
    split_path = split_path_template.format(split=split)
    split_json = load_json(split_path)

    new_json = {}
    for image_id, info in split_json.items():
        caption = info["caption"]
        ocr_description = info["ocr_description"]

        while "_" in caption:
            caption = caption.replace("_", " ")
        while "_" in ocr_description:
            ocr_description = ocr_description.replace("_", " ")

        new_json[image_id] = {
            "caption": caption,
            "ocr_description": ocr_description,
        }
    save_json(save_path_template.format(split=split), new_json)
```
